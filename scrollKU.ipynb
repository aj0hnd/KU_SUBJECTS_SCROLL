```python
from time import sleep 
from openpyxl import Workbook
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException

url = 'https://sugang.korea.ac.kr/'
wb = Workbook(write_only=True)
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_argument('--headless')
options.add_argument('window-size=1920,1080')
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.maximize_window()
wait = WebDriverWait(driver, 20)

driver.switch_to.frame('Main')

sleep(2)
driver.find_element(by=By.CSS_SELECTOR, value='div.jconfirm-closeIcon').click()
sleep(0.5)
driver.find_element(by=By.CSS_SELECTOR, value='#id').send_keys('2022320075')
driver.find_element(by=By.CSS_SELECTOR, value='#pwd').send_keys('naewoo88@1')
driver.find_element(by=By.CSS_SELECTOR, value='#btn-login').click()
sleep(3)

driver.switch_to.frame('coreMain')
print("load..to..table page..")
sleep(3)
wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR,'#menu_hakbu'))).click()
sleep(3)

courdiv_lists = ['00', '24', '01']

sleep(1)

for courdiv in courdiv_lists:
    course_div = Select(driver.find_element(by=By.CSS_SELECTOR, value='#pCourDiv'))
    course_div.select_by_value(courdiv)
    if courdiv != '01':
        if courdiv == '00': 
            ws = wb.create_sheet('전공')
            ws.append(['학수번호', '이수구분', '개설학과', '교과목명', '담당교수', '학점(시간)', '강의시간/강의실'])
        else: 
            ws = wb.create_sheet('학문의 기초')
            ws.append(['학수번호', '이수구분', '개설학과', '교과목명', '담당교수', '학점(시간)', '강의시간/강의실'])
        find_pCols = driver.find_elements(by=By.CSS_SELECTOR, value='#pCol > option')
        sleep(1)
        for pcols in find_pCols:
            col = pcols.text
            col_div = Select(driver.find_element(by=By.CSS_SELECTOR, value='#pCol'))
            sleep(0.5)
            col_div.select_by_visible_text(col)
            sleep(0.5)
            find_pDept = driver.find_elements(by=By.CSS_SELECTOR, value='#pDept > option')
            sleep(1)
            for depts in find_pDept:
                dept = depts.text
                dept_div = Select(driver.find_element(by=By.CSS_SELECTOR, value='#pDept'))
                sleep(0.5)
                dept_div.select_by_visible_text(dept)
                sleep(0.5)
                driver.find_element(by=By.CSS_SELECTOR, value='#btnSearch').click()
                sleep(3)
                all_subjects = driver.find_elements(by=By.CSS_SELECTOR, value='tbody > tr.odd, tbody >  tr.even')
                sleep(0.5)
                for subject in all_subjects:
                    try:
                        courseNum = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(2) > span').text
                        types = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(4)').text
                        major = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(5)').text
                        subject_name = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(6)').text
                        professor = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(7)').text
                        time = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(8)').text
                        location = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(9)').text
                        ws.append([courseNum, types, major, subject_name, professor, time, location])
                        sleep(0.5)
                    except NoSuchElementException:
                        break
                print(f'done : {col}, {dept}')
    else: 
        ws = wb.create_sheet('교양')
        ws.append(['학수번호', '이수구분', '개설학과', '교과목명', '담당교수', '학점(시간)', '강의시간/강의실'])
        find_pGroupCd = driver.find_elements(by=By.CSS_SELECTOR, value='#pGroupCd > option')
        print(find_pGroupCd[0].text)
        sleep(1)
        for groups in find_pGroupCd:
            group = groups.text
            dept_div = Select(driver.find_element(by=By.CSS_SELECTOR, value='#pGroupCd'))
            sleep(0.5)
            dept_div.select_by_visible_text(group)
            sleep(0.5)
            driver.find_element(by=By.CSS_SELECTOR, value='#btnSearch').click()
            sleep(3)
            all_subjects = driver.find_elements(by=By.CSS_SELECTOR, value='tbody > tr.odd, tbody >  tr.even')
            sleep(0.5)
            for subject in all_subjects:
                try:
                    courseNum = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(2) > span').text
                    types = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(4)').text
                    major = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(5)').text
                    subject_name = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(6)').text
                    professor = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(7)').text
                    time = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(8)').text
                    location = subject.find_element(by=By.CSS_SELECTOR, value='td:nth-child(9)').text
                    ws.append([courseNum, types, major, subject_name, professor, time, location])
                    sleep(0.1)
                except NoSuchElementException:
                    break
sleep(2)
driver.quit()
wb.save('/Users/anjonghyeon/Desktop/MeMyselfAndI/KU_SUBJECTS_SCROLL/result/schedule.xlsx')
```
